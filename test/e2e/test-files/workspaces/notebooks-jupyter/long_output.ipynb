# Long Output

## Text Output

```python
from IPython.display import HTML

import pandas as pd
```

Output just short of max height and `notebook.output.textLineLimit`:

- `notebook.output.scrolling` enabled: should not scroll
- `notebook.output.scrolling` disabled: should not truncate

```python
for i in range(29):
    print(i)
```

Output still within `notebook.output.textLineLimit` but just exceeding max height:

- `notebook.output.scrolling` enabled: should scroll - ideally shouldn't scroll but not sure if it's worth the added complexity to handle this case
- `notebook.output.scrolling` disabled: should not truncate

```python
for i in range(30):
    print(i)
```

Output much longer than max height and `notebook.output.textLineLimit`:

- `notebook.output.scrolling` enabled: should scroll
- `notebook.output.scrolling` disabled: should truncate

```python
for i in range(50):
    print(i)
```

## Non-Text Output

Long non-text output:

- `notebook.output.scrolling` enabled: should scroll
- `notebook.output.scrolling` disabled: should not truncate - there's no obvious truncation point, but we hope that third party libraries handle truncation

```python
with pd.option_context('display.max_rows', None):
    display(HTML(pd.DataFrame({'a': range(40)})._repr_html_()))
```

Wide non-text output, should horizontally scroll:

```python
with pd.option_context('display.max_columns', None):
    display(HTML(pd.DataFrame({'my long column name '+str(k): [0] for k in range(100)})._repr_html_()))
```

Wide and long non-text output:

- Should horizontally scroll
- `notebook.output.scrolling` enabled: should scroll
- `notebook.output.scrolling` disabled: should not truncate - there's no obvious truncation point, but we hope that third party libraries handle truncation

```python
with pd.option_context('display.max_rows', None):
    display(HTML(pd.DataFrame({'my long column name '+str(k): range(50) for k in range(1000)})._repr_html_()))
```
